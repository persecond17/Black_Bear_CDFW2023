```python
bears = pd.read_csv('data/final_df_copy.csv')
bears['dist_from_campsite'] = ''
```

```python
for i in range(bears.shape[0]):
    bear_point = gpd.points_from_xy([bears.iloc[i,0]], [bears.iloc[i,1]])
    dist = 1000000
    for j in range(df.shape[0]):
        dist = min(dist, df.iloc[j,-1].distance(bear_point[0]))
    bears.iloc[i,-1] = dist * 0.000621 # convert to miles
    # bears.iloc[i,-1] = dist
```

```python
df = pd.read_csv('data/inat/inat_04252023.csv')
df1 = df.iloc[0:5,:].copy()
df1
```

```python
df1 = df1[['latitude','longitude']]
df1
```

```python
from tqdm import tqdm
from geopy.geocoders import Nominatim
```

```python
import time
```

```python
geolocator = Nominatim(user_agent="cdfw-model")
county = []
geocoords = df1[['latitude', 'longitude']].values
for elems in tqdm(geocoords):
    tracker = False
    address, _ = geolocator.reverse((f'{elems[0]}, {elems[1]}'))
    address = address.split(',')
    for i in address:
        if 'county' in i.lower():
            tracker = True
            county_name = i.strip().strip('County').strip()
            continue
    
    if tracker:
        county.append(county_name)
    else:
        county.append('')
    
    time.sleep(0.5)

print(county)
df1 = pd.concat([df1, pd.Series(county, name='county')], axis=1)
```
