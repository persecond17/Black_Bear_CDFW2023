### Calculate distance from bear sighting to nearest water

```python
bears['dist_from_water'] = ''
```

```python
for i in range(bears.shape[0]):
    bear_point = gpd.points_from_xy([bears.iloc[i,1]], [bears.iloc[i,2]])
    county_df_name = 'df_' + bears.iloc[i,-2].replace(' ', '')
    dist = 1000000
    if county_df_name in dfs.keys():
        for j in range(dfs[county_df_name].shape[0]):
            dist = min(dist, df.iloc[j,-1].distance(bear_point[0]))
    else:
        dist = min(dist, df.iloc[i,-1].distance(bear_point[0]))

    bears.iloc[i,-1] = round(dist * 0.000621) # convert to miles
```

# Add landcode

```python
def dist_btw_2_coords(coord1, coord2):
    """
    Calculates the Euclidean distance between two coordinate points.
    """
    from math import sin, cos, sqrt, atan2, radians
    
    # Approximate radius of earth in km
    R = 6373.0


    lat1 = radians(coord1[0])
    lon1 = radians(coord1[1])
    lat2 = radians(coord2[0])
    lon2 = radians(coord2[1])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance_in_miles = R * c * 0.621371

    # print("Result: ", distance_in_miles)
    
    return distance_in_miles
```

```python
df_nlcd = pd.read_csv('data/nlcd_final.csv')

bears['landcode'] = '99'

for i in range(bears.shape[0]):
    bear_coord = bears.iloc[i, 1:3].values
    bear_lat = bear_coord[0]
    bear_lon = bear_coord[1]
    temp_nlcd = df_nlcd.copy()
    temp_nlcd = temp_nlcd[temp_nlcd['lat'].between(bear_lat-2, bear_lat+2)]
    temp_nlcd = temp_nlcd[temp_nlcd['lon'].between(bear_lon-2, bear_lon+2)]
    temp_nlcd['dist'] = temp_nlcd.apply(lambda x: dist_btw_2_coords(bear_coord, [x['lat'], x['lon']]), axis=1)
    code = temp_nlcd.sort_values('dist').head(1).iloc[:, 2].values[0]
    bears.iloc[i, -1] = code

bears.head()
```

```python
bears.to_csv('data/inat/inat_04252023_final.csv', index=False)
```
